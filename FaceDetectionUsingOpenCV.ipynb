```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
prasanna = cv2.imread('../DATA/Me.jpg',0)
office_team = cv2.imread('../DATA/TreatDay.jpg',0)
school = cv2.imread('../DATA/School.jpg',0)
```

```python
plt.imshow(school,cmap='gray')
```

```python
# face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
face_cascade = cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detect_face(img):
    
  
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
    
```

```python
result = detect_face(office_team)
```

```python
plt.imshow(result,cmap='gray')
```

```python
Enhancing the face recognition by using scaleFactor and minimum neighbor
```

```python
def adj_detect_face(img):
    
    face_img = img.copy()
  
    face_rects = face_cascade.detectMultiScale(face_img,scaleFactor=1.2, minNeighbors=5) 
    
    for (x,y,w,h) in face_rects: 
        cv2.rectangle(face_img, (x,y), (x+w,y+h), (255,255,255), 10) 
        
    return face_img
```

```python
result = adj_detect_face(school)
```

```python
plt.imshow(result,cmap = 'gray')
```

```python
# FACE DETECTION IN VIDEO
```

```python
cap = cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    
    frame = detect_face(frame)
    
    cv2.imshow('Video face detect', frame)
    
    k = cv2.waitKey(1)
    if k ==27:
        break
        

cap.release()
cv2.destroyAllWindows()
```
